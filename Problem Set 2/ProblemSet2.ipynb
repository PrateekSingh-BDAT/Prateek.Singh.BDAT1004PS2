Write a class named Marsupial that can be used as shown below:

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.
k = Kangaroo(0,0)
print(k)
I am a Kangaroo located at coordinates (0,0)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
['doll', 'firetruck', 'kitten']
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
I am a Kangaroo located at coordinates (3,0)

```python
#Answer 3:

class Marsupial: 
    'Defining a class marsupial'
    def __init__(self, xcoord=0, ycoord=0): #constructor function
        self.x = xcoord 
        self.y = ycoord
        self.pouch = []
    
    def jump(self, dx, dy):
        self.x = self.x + dx
        self.y = self.y + dy
        
    def put_in_pouch(self,pouchContent):
        self.pouch.append(pouchContent)
    
    def pouch_contents(self):
        print(self.pouch)
        
    def __str__(self): # canonical representation function
        return "I am a marsupial located at coordinates ({},{})".format(self.x,self.y)
        
class kangaroo(Marsupial):
    'defining class for marsupial kangaroo'
    def __init__(self,xcoord=0, ycoord=0):
        Marsupial.__init__(self,xcoord,ycoord) #extending method __init__ from super class.
    def __str__(self):
        return  "I am a Kangaroo located at coordinates ({},{})".format(self.x,self.y) #overloding method __str__ from super class
    
k = kangaroo(0,0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

## Question5
Write a recursive method binary() that takes a non-negative
integer n and prints the binary representation of integer n.

binary(0)
0
binary(1)
1
binary(3)
11
binary(9)
1001

```python
#Answer 5
binaryVal = ""
def binary(decimal):
        global binaryVal 
        if (decimal==1):
            binaryVal = str(decimal) + binaryVal 
        elif (decimal == 0):
            print(decimal)
        else:
            binaryVal = str(decimal%2) + binaryVal 
            binary(decimal//2) #recursive call
        print(binaryVal)
        binaryVal = ""
binary(9)
```

Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. Your function should visit every web
page reachable from the starting URL web page in depth clicks or less, and print
each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL.

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html

 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

```python
#Answer 7

from urllib.request import urlopen
from urllib.parse import urljoin
from html.parser import HTMLParser

class Collector(HTMLParser):
    'collects hyperlink URLs into a list'

    def __init__(self, url):
        'initializes parser, the url, and a list'
        HTMLParser.__init__(self)
        self.url = url
        self.links = []

    
    def handle_starttag(self, tag, attrs):
        'collects hyperlink URLs in their absolute format'
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    # construct absolute URL
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http': # collect HTTP URLs
                        self.links.append(absolute)
                        
    def getLinks(self):
        'returns hyperlinks URLs in their absolute format'
        return self.links

def webdir(URL,depth,indent):
    'creating the recursive function'
    resource = urlopen(URL)
    content = resource.read().decode()
    collector = Collector(URL)
    collector.feed(content)
    linksList = collector.getLinks()
    indentIndex = 0
    level = 0 #to indent the URLs we need to set a variable level to specify the level of the URL
    if (depth == 0):
        for indentIndex in range(0,len(linksList),1):
            print(linksList[indent])
    else:
        while (level<=depth):
            for indentIndex in range(0,len(linksList),1):
                print(" "*level, linksList[indentIndex])
                level = level + 1
            
    linkIndex = 0
    while (depth != 0):
        for linkIndex in range(0,len(linksList),1):
            webdir(linksList[linkIndex],depth,indent) #recursive call
        depth = depth - 1
        
web = webdir('https://reed.cs.depaul.edu/lperkovic/test1.html', 2, 0)
```
